```python
def show_similar_posts(df, similarity_threshold=90):
    """
    Return a DataFrame listing all pairs of posts (by index) that are similar 
    above the given similarity_threshold. It uses the 'Normalized' column.
    """
    similar_pairs = []

    # We'll compare each row with all subsequent rows
    for i in range(len(df)):
        for j in range(i + 1, len(df)):
            # Compare the normalized texts using fuzzy ratio
            ratio = fuzz.ratio(df.iloc[i]['Normalized'], df.iloc[j]['Normalized'])
            if ratio >= similarity_threshold:
                similar_pairs.append({
                    'Index_1': df.index[i],
                    'Text_1': df.iloc[i]['Normalized'],
                    'Index_2': df.index[j],
                    'Text_2': df.iloc[j]['Normalized'],
                    'SimilarityRatio': ratio
                })

    # Convert to a DataFrame for easy viewing
    return pd.DataFrame(similar_pairs)

def normalize_post(text):
    """
    Convert the text to lowercase and remove URLs, hashtags, mentions, punctuation,
    and extra whitespace.
    """
    if not isinstance(text, str):
        # Convert non-string (like float NaN) to an empty string
        text = ""
    text = text.lower()
    text = re.sub(r'http\S+', '', text)   # Remove URLs
    text = re.sub(r'[@#]\S+', '', text)     # Remove hashtags and mentions
    #text = re.sub(r'[^\w\s]', '', text)     # Remove punctuation
    return text.strip()
```
