```python

timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
remove_terms = {
        'help', 'get', 'twitter', 'like', 'one', 'using', 'see',
        'youtube', 'today', 'check', 'us', 'made', 'first', 'time', 'learn', 'de', 'want', 'first',
        'week', 'know', 'good', 'year', 'day', 'need', 'project', 'would','high','free', 'think',
        'utm_medium', 'utm_source', 'igshid',
        'post','way','also', 'thanks',
        'something','coming','utm_source twitter','years',
        'two', 'look', 'going', 'latest', 'got', 'take', 'utm_campaign', 'x', 
        'come', 'another', 'hours', 'really', 'utm_source twitter', 'youtube video',

        # generic filler / low-value
        'work', 'heart', 'read', 'people', 'set', 'based', 'designed',
        'social', 'building', 'go', 'best', 'full', 'team', 'life',
        'ready', 'buy', 'live', 'watch', 'use', 'working', 'things',
        'love', 'stuff', 'thing', 'back', 'even', 'many', 'make',
        'build', 'find', 'start', 'last', 'ever', 'test', 'share',
        'new', 'r', 'e', 'w', 'hp', 'drone', 'app', 'twitter',

        #
        'utm_source twitter', 'utm_source dlvr utm_medium','utm_source dlvr',
        'utm_medium twitter', 'youtube video', 'dlvr utm_medium twitter', 
        'dlvr utm_medium', 'liked youtube', 'liked youtube video', 
        'promo code', 'twitter utm_medium social', 'buy 3d printer', 
        'social utm_campaign social', 'utm_campaign social', 'looks like', 
        'twitter utm_medium', 'learn 3d printing', 'video youtube', 'need 3d', 'utm_source twitter utm_medium', 'utm_medium social',
        'first ever','see 3d', 'utm_source twitter 3d',
        'join us','via youtube','like 3d', 'twitter 3d printing', '3d 3d', 'like 3d printing', 'buy 3d', 'men women',
        'world 3d', 'could 3d', 'printing one', 'thanks 3d', 'fully 3d', 'love 3d', 'using 3d', 'first time', 'learn 3d',

        # 'utm_medium rss utm_campaign',
        'utm_source rss utm_medium', 'rss utm_medium rss', 'printing news unpeeled', 'video youtube playlist', 'added video youtube',
        'youtube video 3d', 'youtube 3d printing', 'html channel twinner', 'ttranpham azul3d com', 'get 3d printing',
        'got 3d printer', 'get 3d printer', 'using 3d printing', 'printed 3d printing', 'print 3d printing','3d printing going',
        'see 3d printing', '3d printing 3d', 'printing 3d printing','3dp 3d printing','3d printing one','22 20or 20',
        'filament 1 75mm',
        'utm_medium rss utm_campaign',
        '3d printing html',
        'working 3d printing',
        'latest additive manufacturing',
        '3d printing tuan',
        'acturing 3d printing',
        '3d 3d printing',
        'latest 3d printing',
        '3d printing project',
        '3d printing world',
        'unveils 3d printing', 'bring 3d printing', '3d printing 3dmodeling', '3d printing html', 'latest additive manufacturing', 'love 3d printing'
        '3d printing shirt', 'help 3d printing', '3d printer review', 'first 3d printing', '3d printing workshop', 'today 3d printing', 
        'know 3d printing', 'learning 3d printing',
        'utm_medium social utm_campaign', '3d printing case', 'new 3d printing',
        'got 3d printing', 'additive manufacturing news', 'made 3d printing',
        'created 3d printing', 'one 3d printing', '3d printing post',
        'uses 3d printer', 'love 3d printing', 'want 3d printing',
        '3d printing shirt', '3d printing using',
        'utm_medium social utm_campaign',
        'social utm_campaign',

        # unigram specific
        "still", "fun", "game", "looking", "available", "models", "com", "used", "kit", "right", 
        "getting", "students", "anyone", "school", "little", "top", "story", "looks", "thank", "man", "great", "hand", "soon",
        "let", "liked", "big", "done", "small", "show", "support", "print", "pro", "student", "3dp",
        'en', 'better','size','stem','human','much','well','large','learning','link','creality','create',
        

        # quad grams
        "added video youtube playlist", "liked youtube video 3d", "19255 0 1 ff3", "artemis media tw ebay",
        'excited share latest addition', 'share latest addition etsy', 'widget_ver artemis media tw',
        'added video youtube playlist',
        "youtube video 3d printing",
        "printing 3d printing 3d","3d printing 3ders twitter","acturing 3d printing news","19255 0 1 ff3","0 1 ff3 2",
        "1 ff3 2 toolid","2 toolid 10039 campid","ff3 2 toolid 10039","4f55 855d fc12069811f7 posts","00a7 4f55 855d fc12069811f7",
        "see ews 3d printing", "uses 3d printer make","feature share via youtube","excited share latest addition","share latest addition etsy",
        "artemis media tw ebay","widget_ver artemis media tw","gostei de um youtube","youtube playlist 3d printing","127632 2357 0 ssspo",
        "1 mkrid 711 127632","latest addition etsy shop","troadeyinc 3d printing quebec","16 mkevt 1 mkrid","3d printing news latest",
        "printing news latest additive","news latest additive manufacturing","video youtube playlist 3d","printer join us discord",
        "3d printing quebec canada","3d printing top news","3d models 3d printing","711 127632 2357 0","mkrid 711 127632 2357",
        "3dable 3dp 3d printing", "3d printing iron man",
        "3d printing save icon","use 3d printer make","3d printing thing saw","3d printing maker news","printing thing saw today",
        "printing maker future news","best 3d printing thing","printing 3d printing maker","3d printing industry news",
        "pla filament 1 75mm","printing technology 3d printing","ustado un de youtube","news metal additive manufacturing",
        "3d printing baby yoda","3d printing news briefs","social sh soc_trk tw",
        "soc_src social sh soc_trk","200 19255 0 1","3d printing news unpeeled","ich mag das youtube","mag das youtube video",
        "get started 3d printing","3d printing 3d 3d","printing parts 3d printing","un video di youtube","piaciuto un video di",
        "mi piaciuto un video","printing 3d 3d printing","j aime une youtube",
        "printer 3d printing additive","additive manufacturing users group",
        "polymer additive manufacturing news","news polymer additive manufacturing",
        "printing news polymer additive","3d printing news polymer","printing 3d 3d printing","22 20or 20 223","gustado un de youtube",
        "22 20or 20 223d"

}

term_map_quad_grams = {
    'first 3d printing gun': '3d printing guns related',
    '3d printing gun blueprints': '3d printing guns related',
    '3d printing ghost gun': '3d printing guns related',
    '3d printing gun files': '3d printing guns related',
    'printing gun 3d printing': '3d printing guns related',
    '3d printing gun 3d': '3d printing guns related',
    '3d printing plastic gun': '3d printing guns related',
    'entirely 3d printing gun': '3d printing guns related',
    '3d printing metal gu': '3d printing guns related',
    'fully 3d printing gun': '3d printing guns related',
    'selling 3d printing gun': '3d printing guns related',
    'printing gun fired video': '3d printing guns related',
    '3d printing gun terrifying': '3d printing guns related',
    'printing next frontier gun': '3d printing guns related',
}

def standardize_ngrams(df, term_map):
    df = df.copy()
    for variant, standard in term_map.items():
        df['ngram'] = df['ngram'].str.replace(
            r'\b' + re.escape(variant) + r'\b', 
            standard, 
            regex=True, 
            flags=re.IGNORECASE
        )
    return df

# ───────────────────────────────
def normalize_variants(df):
    fixes = {
        r'\bresearch(?:ers)?\b':                   'research',
        r'\bmodels?\b':                           'models',
        r'\bparts?\b':                            'parts',
        r'\b3d[-\s]*print(?:ing|ed|ted)?\b':      '3d printing',
        r'\b3dprinting\b':                        '3d printing',
        r'\b3d[-\s]*printer(?:s)?\b':             '3d printer',
        r'\badditive[-\s]*manufactur(?:e|ed|ing)?\b': 'additive manufacturing',
        r'\bmaterials?\b':                        'material',
        r'\brapid[-\s]*prototyp(?:e|ed|ing)?\b':  'rapid prototyping',
        r'\brobots?\b':                           'robotics',
        r'\b3dprint(?:industry|industries)?\b': '3d printing industry',
        # convert singular or plural “gun” to “guns”
        r'\bguns?\b':                              'gun',
        r'\bdesigns?\b':                           'design',
        r'\bapplications?\b':                      'application',
        r'\bscientists?\b':                        'science',
        r'\bindustrial\b':                         'industry',
        r'\bcompanies?\b':                         'business',
    }
    text = df['Normalized Text'].fillna('')
    for patt, repl in fixes.items():
        text = text.str.replace(patt, repl, regex=True, flags=re.IGNORECASE)
    text = text.str.replace(r'\b(\w+)(?:\s+\1\b)+', r'\1', regex=True, flags=re.IGNORECASE)
    out = df.copy()
    out['Normalized Text'] = text
    return out

# ───────────────────────────────
# 2. Extract binary n-grams + compute PMI
# ───────────────────────────────
def extract_year_ngrams_with_pmi(df, custom_stopwords,
                                 n_range=(2,3), min_df=2, top_n=200):
    texts = df['Normalized Text'].fillna('').tolist()
    total_docs = len(texts)
    if total_docs == 0:
        return pd.DataFrame(columns=['ngram','frequency','pmi'])

    # document-frequency for n-grams
    vec_ng = CountVectorizer(
        ngram_range=n_range,
        stop_words=list(custom_stopwords),
        min_df=min_df, # drop any n-gram that appears in fewer than min_df tweets...minimum document frequency
        token_pattern=r'\b[a-zA-Z0-9_]+\b',
        binary=True # record presence/absence per tweet, not raw counts per tweet
    )
    X_ng = vec_ng.fit_transform(texts)
    ngrams = vec_ng.get_feature_names_out()
    freq_ng = X_ng.sum(axis=0).A1

    # document-frequency for unigrams (for PMI)
    vec_uni = CountVectorizer(
        ngram_range=(1,1),
        stop_words=list(custom_stopwords),
        token_pattern=r'\b[a-zA-Z0-9_]+\b',
        binary=True
    )
    X_uni = vec_uni.fit_transform(texts)
    unigrams = vec_uni.get_feature_names_out()
    freq_uni = X_uni.sum(axis=0).A1
    uni_index = {w: i for i, w in enumerate(unigrams)}

    # build DataFrame with PMI
    records = []
    for i, term in enumerate(ngrams):
        count_ng = freq_ng[i]
        words = term.split()
        try:
            p_ng = count_ng / total_docs
            p_words = 1.0
            for w in words:
                idx = uni_index.get(w)
                count_w = freq_uni[idx] if idx is not None else 0
                p_words *= (count_w / total_docs)
            pmi = math.log2(p_ng / p_words) if p_ng > 0 and p_words > 0 else float('-inf')
        except Exception:
            pmi = float('-inf')
        records.append({'ngram': term, 'frequency': count_ng, 'pmi': pmi})

    df_ngrams = pd.DataFrame(records)
    return df_ngrams.sort_values('frequency', ascending=False).head(top_n)

# ───────────────────────────────
# 3. Find trending terms (including PMI)
# ───────────────────────────────
def find_trending_terms(all_ngrams_df, remove_terms, min_years=2, top_n=15):

    # ──────────────────────────────────────────────────────────────────
    # 1) Remove unwanted n-grams (your custom stop-list + pure digits)
    # ──────────────────────────────────────────────────────────────────
    df = all_ngrams_df[
        (~all_ngrams_df['ngram'].isin(remove_terms)) &
        (~all_ngrams_df['ngram'].str.isdigit())
    ]

    # ──────────────────────────────────────────────────────────────────
    # 2) Require each n-gram to appear in at least `min_years` distinct years
    # ──────────────────────────────────────────────────────────────────
    #    This ensures we’re looking at multi-year trends, not one-offs.
    common = df.groupby('ngram').filter(lambda g: g['year'].nunique() >= min_years)

    # ──────────────────────────────────────────────────────────────────
    # 3) Compute per-term metrics (absolute & relative change, PMI, etc.)
    # ──────────────────────────────────────────────────────────────────
    records = []
    for term, grp in common.groupby('ngram'):
         # sort by year so we can compare “first” vs. “last”
        grp = grp.sort_values('year')
        first, last = grp.iloc[0], grp.iloc[-1] # ontain first and last year

        # how much did frequency change (per 100 tweets)?
        abs_ch = last['freq_per_100_tweets'] - first['freq_per_100_tweets']

        #######################################
        if abs_ch <= 0:
            continue
        #######################################

        # relative change = (last − first) / first
        # with (guard against division by zero)
        rel_ch = abs_ch / first['freq_per_100_tweets'] if first['freq_per_100_tweets'] > 0 else float('inf')

        
        # include PMI from the last year occurrence
        pmi_val = float(last.get('pmi', float('nan')))
        records.append({
            'term': term,
            'first_year': int(first['year']),
            'last_year': int(last['year']),
            'first_freq': first['freq_per_100_tweets'],
            'last_freq': last['freq_per_100_tweets'],
            'abs_change': abs_ch,
            'rel_change': rel_ch,
            'pmi': pmi_val,
            'word_count': int(first['word_count'])
        })

    return (
        pd.DataFrame(records)
          .sort_values('rel_change', ascending=False)
          .head(top_n)
    )

# ───────────────────────────────
# 4. One-time setup & run
# ───────────────────────────────
pd.set_option("display.min_rows", 50)

nltk.download('stopwords')
nltk.download('punkt')
stops = set(stopwords.words('english'))
twitter_stops = {'rt','amp','http','https','co','t','s','m','re','ve','ll','d'}
custom_stops = stops.union(twitter_stops)

# Load & prep data
df = pd.read_csv('CLEANED_DATASET_20250531_004116.csv', parse_dates=['Date'])
df['year'] = df['Date'].dt.year
if 'Normalized Text' not in df:
    df['Normalized Text'] = df['Text']

# Normalize text
df = normalize_variants(df)

# Build per-year binary n-grams with PMI
results = []
for yr, sub in df.groupby('year'):
    if len(sub) < 5:
        continue
    ngr = extract_year_ngrams_with_pmi(sub, custom_stops, n_range=(4,4), min_df=5, top_n=200)
    ngr['year'] = yr
    ngr['tweet_count'] = len(sub)
    ngr['freq_per_100_tweets'] = 100 * ngr['frequency'] / len(sub)
    ngr['word_count'] = ngr['ngram'].str.count(' ') + 1
    results.append(ngr)

###--- Standardization of terms

all_ngrams = pd.concat(results, ignore_index=True)
#quad
all_ngrams = standardize_ngrams(all_ngrams, term_map_quad_grams)

# Recalculate word count after standardizing
all_ngrams['word_count'] = all_ngrams['ngram'].str.count(' ') + 1

all_ngrams = (
    all_ngrams
    .groupby(['ngram', 'year'], as_index=False)
    .agg({
        'frequency': 'sum',
        'pmi': 'first',  # or 'last' or another strategy
        'tweet_count': 'sum',
        'freq_per_100_tweets': 'sum',
        'word_count': 'first'
    })
)
#####

# Compute & print trending terms
trending = find_trending_terms(all_ngrams, remove_terms, min_years=2, top_n=200)
trending.to_csv(f'trending_terms_{timestamp}.csv', index=False)
trending
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_term_heatmap(terms, all_ngrams_df, normalize=True, cmap='YlGnBu'):
    """
    Plot a heatmap of (optionally normalized) frequencies of terms over time.

    Args:
        terms (list): List of terms (n-grams) to visualize.
        all_ngrams_df (DataFrame): DataFrame with n-gram frequencies and year.
        normalize (bool): Whether to normalize frequencies per term.
        cmap (str): Colormap for heatmap.

    Returns:
        None: Displays a heatmap plot.
    """
    df = all_ngrams_df[all_ngrams_df['ngram'].isin(terms)]
    pivot = df.pivot(index='ngram', columns='year', values='freq_per_100_tweets')

    if normalize:
        pivot = pivot.div(pivot.max(axis=1), axis=0)  # Normalize each row by its max

    plt.figure(figsize=(12, len(terms) * 0.6 + 2))
    sns.heatmap(pivot.fillna(0), annot=True, fmt=".2f", cmap=cmap, linewidths=0.5)
    plt.title('Term Frequency Over Time (Normalized per Term)' if normalize else 'Term Frequency Over Time')
    plt.xlabel('Year')
    plt.ylabel('Term')
    plt.tight_layout()
    plt.show()
```

```python
def filter_trends_by_terms(trend_df, terms, term_col='term'):
    # ensure uniqueness & preserve order
    seen = set()
    ordered = [t for t in terms if not (t in seen or seen.add(t))]

    # filter
    sub = trend_df[trend_df[term_col].isin(ordered)].copy()

    # reindex to match the order of your input list
    sub[term_col] = pd.Categorical(sub[term_col], categories=ordered, ordered=True)
    return sub.sort_values(term_col)

# Example usage:
terms_of_interest = ['rocket', 'material', 'fun', 'covid']
subset = filter_trends_by_terms(trending, terms_of_interest)
subset
```

```python
def filter_ngrams(df: pd.DataFrame,
                  year: int,
                  min_pmi: float,
                  min_freq_per_100: float) -> pd.DataFrame:
    # build the boolean mask:
    mask = (
        (df['year'] == year) &
        (df['pmi'] >= min_pmi) &
        (df['freq_per_100_tweets'] >= min_freq_per_100)
    )
    return df.loc[mask].copy()
# suppose `all_ngrams` is your DataFrame
filtered = filter_ngrams(all_ngrams, year=2012, min_pmi=1.0, min_freq_per_100=2.0)
filtered
```

```python
def plot_dumbbell(df, picks, 
                  term_col='term',
                  start_col='first_freq', end_col='last_freq',
                  start_label='First year freq', end_label='Last year freq',
                  width=6, height_per_term=0.8):
    """
    df              : DataFrame containing your trend_df
    picks           : list of integer indices or list of term-values
    width           : total figure width in inches
    height_per_term : height in inches for each term/row
    """
    # subset & sort
    if all(isinstance(p, int) for p in picks):
        sub = df.loc[picks].copy()
    else:
        sub = df[df[term_col].isin(picks)].copy()
    sub = sub.sort_values(by=start_col)
    
    # dynamic figure size
    n = len(sub)
    figsize = (width, max(4, height_per_term * n))
    fig, ax = plt.subplots(figsize=figsize)
    
    # positions
    y = range(n)
    
    # grid, lines, markers
    ax.grid(axis='x', linestyle='--', alpha=0.5)
    ax.hlines(y, sub[start_col], sub[end_col], color='gray', linewidth=1)
    ax.scatter(sub[start_col], y, label=start_label, zorder=3)
    ax.scatter(sub[end_col],   y, label=end_label,   zorder=3)
    
    # annotate years dynamically if you want...
    
    # labels & ticks
    ax.set_yticks(y)
    ax.set_yticklabels(sub[term_col])
    ax.set_xlabel('Frequency per 100 tweets')
    ax.legend(loc='upper right', frameon=False)
    ax.invert_yaxis()
    
    # margins
    plt.subplots_adjust(right=0.85, left=0.25)
    plt.tight_layout()
    return fig, ax
```

```python
from math import isfinite

def find_peaking_terms(all_ngrams_df, min_years=3, top_n=20):
    """
    Finds ngrams that have a clear local max (peak) and then decline.
    Returns the terms with the largest relative drop from peak to last year.
    """
    records = []
    # only consider terms that appear in at least min_years
    grouped = all_ngrams_df.groupby('ngram').filter(lambda g: g['year'].nunique() >= min_years)
    for term, grp in grouped.groupby('ngram'):
        # sort by year
        grp = grp.sort_values('year')
        years = grp['year'].values
        freqs = grp['freq_per_100_tweets'].values

        # find all local maxima (ignore flat plateaus)
        peak_idxs = [i for i in range(1, len(freqs)-1)
                     if freqs[i] > freqs[i-1] and freqs[i] > freqs[i+1]]
        if not peak_idxs:
            continue

        # pick the strongest peak
        peak_i = peak_idxs[freqs[peak_idxs].argmax()]
        peak_year = int(years[peak_i])
        peak_freq = freqs[peak_i]

        # how much does it drop from that peak to the last year?
        last_freq = freqs[-1]
        abs_drop = peak_freq - last_freq
        rel_drop = abs_drop / peak_freq if peak_freq > 0 else float('inf')
        if not isfinite(rel_drop) or rel_drop <= 0:
            continue

        records.append({
            'term': term,
            'peak_year': peak_year,
            'peak_freq': peak_freq,
            'last_year': int(years[-1]),
            'last_freq': last_freq,
            'abs_drop': abs_drop,
            'rel_drop': rel_drop,
            'word_count': int(term.count(' ') + 1)
        })

    return (pd.DataFrame(records)
              .sort_values('rel_drop', ascending=False)
              .head(top_n))

# usage:
peaks = find_peaking_terms(all_ngrams, min_years=3, top_n=1500)
peaks
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def build_three_point_df(trending_df, peaks_df, terms):
    # make a lookup of peak info
    peak_info = peaks_df.set_index('term')[['peak_year','peak_freq']]

    valid_terms = [t for t in terms if t in peak_info.index]
    missing = set(terms) - set(valid_terms)
    if missing:
        print(f"⚠️ Skipping these (no peak detected): {sorted(missing)}")

    rows = []
    for term in valid_terms:
        # first & last from your trending DF
        trow = trending_df[trending_df['term']==term].iloc[0]
        fy, ff = trow['first_year'], trow['first_freq']
        ly, lf = trow['last_year'],  trow['last_freq']
        py, pf = peak_info.loc[term, ['peak_year','peak_freq']]

        rows += [
            {'term':term, 'year': fy, 'freq': ff, 'stage':'first'},
            {'term':term, 'year': py, 'freq': pf, 'stage':'peak'},
            {'term':term, 'year': ly, 'freq': lf, 'stage':'last'},
        ]
    return pd.DataFrame(rows), valid_terms

def plot_three_points(df3, terms):
    fig, ax = plt.subplots(figsize=(8, 0.6*len(terms) + 2))
    for i, term in enumerate(terms):
        sub = df3[df3['term']==term].sort_values('year')
        ax.plot(sub['year'], sub['freq'], marker='o')
        # annotate peak year
        peak = sub[sub['stage']=='peak'].iloc[0]
        ax.text(peak['year'], peak['freq'] + 0.1,
                str(int(peak['year'])),
                ha='center', va='bottom', fontsize='x-small')

    ax.set_yticks([])
    ax.set_yticklabels([])
    ax.set_ylabel('Freq per 100 tweets')
    ax.set_xlabel('Year')
    ax.set_title('First → Peak → Last frequencies')
    # put terms on the right as labels
    for i, term in enumerate(terms):
        last = df3[(df3.term==term)&(df3.stage=='last')].iloc[0]
        ax.text(last['year'] + 0.2, last['freq'],
                term, va='center', fontsize='small')
    plt.tight_layout()
    plt.show()

# --- Usage:

# 1) pick your top N trending terms
top_terms = trending.head(12)['term'].tolist()

# 2) build the 3-point DataFrame (and get valid_terms)
df3, valid_terms = build_three_point_df(trending, peaks, top_terms)

# 3) plot
plot_three_points(df3, valid_terms)
```

```python
def plot_dumbbell_with_peakyear(
    df, picks,
    peak_years,              # dict/Series mapping term → peak_year
    term_col='term',
    start_col='first_freq', end_col='last_freq',
    start_label='First year freq', end_label='Last year freq',
    width=6, height_per_term=0.8
):
    # subset & sort (unchanged)
    if all(isinstance(p, int) for p in picks):
        sub = df.loc[picks].copy()
    else:
        sub = df[df[term_col].isin(picks)].copy()
    sub = sub.sort_values(by=start_col)

    # figure
    n = len(sub)
    fig, ax = plt.subplots(figsize=(width, max(4, height_per_term * n)))

    # positions
    y = range(n)

    # draw the lines & end-points
    ax.grid(axis='x', linestyle='--', alpha=0.5)
    ax.hlines(y, sub[start_col], sub[end_col], color='gray', linewidth=1)
    ax.scatter(sub[start_col], y, label=start_label, zorder=3)
    ax.scatter(sub[end_col],   y, label=end_label,   zorder=3)

    # **new: annotate peak-year at midpoint**
    for i, row in sub.iterrows():
        term = row[term_col]
        if term in peak_years:
            xm = (row[start_col] + row[end_col]) / 2
            ax.text(
                xm, i + 0.1,               # slightly above the line
                str(int(peak_years[term])),
                ha='center', va='bottom',
                fontsize='x-small', color='black'
            )

    # labels & ticks (unchanged)
    ax.set_yticks(y)
    ax.set_yticklabels(sub[term_col])
    ax.set_xlabel('Frequency per 100 tweets')
    ax.legend(loc='upper right', frameon=False)
    ax.invert_yaxis()
    plt.subplots_adjust(right=0.85, left=0.25)
    plt.tight_layout()
    return fig, ax
```

```python
from adjustText import adjust_text

def plot_slope_chart_with_repulsion(df3, terms):
    # … build your 'slope' DataFrame exactly as before …
    pivot = df3.pivot(index='term', columns='stage', values='freq')
    slope = (
        pivot.reset_index()
             .melt(id_vars='term', value_vars=['first','peak','last'],
                   var_name='stage', value_name='freq')
    )
    # order 'stage' and plot
    stage_order = ['first','peak','last']
    slope['stage'] = pd.Categorical(slope['stage'],
                                    categories=stage_order,
                                    ordered=True)
    slope = slope.sort_values(['term','stage'])
    fig, ax = plt.subplots(figsize=(8, 0.6*len(terms) + 2))
    for term, grp in slope.groupby('term'):
        ax.plot(grp['stage'], grp['freq'], marker='o', alpha=0.6)

    # annotate raw freq on each point
    for _, row in slope.iterrows():
        ax.text(row['stage'], row['freq'],
                f"{row['freq']:.1f}",
                va='bottom', ha='center', fontsize='x-small')

    # build label texts only at the last stage
    last_pts = slope[slope['stage']=='last']
    texts = []
    for _, row in last_pts.iterrows():
        # we'll place all at x='last' initially
        texts.append(
            ax.text(
                'last', row['freq'],
                row['term'],
                ha='left', va='center',
                fontsize='small'
            )
        )

    # now push them apart in y, draw leader lines
    adjust_text(texts,
                only_move={'points':'y','text':'y'},
                arrowprops=dict(arrowstyle='-', color='0.5', lw=0.5))

    ax.set_xlabel('Stage')
    ax.set_ylabel('Freq per 100 tweets')
    ax.set_title('Slope Chart with Repelled Labels')
    ax.set_xticks(stage_order)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd

# 1) your existing helper to get the 3-point DataFrame
def build_three_point_df(trending_df, peaks_df, terms):
    peak_info = peaks_df.set_index('term')[['peak_year','peak_freq']]
    valid_terms = [t for t in terms if t in peak_info.index]
    rows = []
    for term in valid_terms:
        trow = trending_df[trending_df['term']==term].iloc[0]
        fy, ff = trow['first_year'], trow['first_freq']
        ly, lf = trow['last_year'],  trow['last_freq']
        py, pf = peak_info.loc[term, ['peak_year','peak_freq']]
        rows += [
            {'term':term, 'stage':'first', 'year':fy, 'freq':ff},
            {'term':term, 'stage':'peak',  'year':py, 'freq':pf},
            {'term':term, 'stage':'last',  'year':ly, 'freq':lf},
        ]
    return pd.DataFrame(rows), valid_terms

# 2) sparkline helper mapping values → ▁▂▃▄▅▆▇█
def sparkline(vals, ticks='▁▂▃▄▅▆▇█'):
    mn, mx = min(vals), max(vals)
    span = mx - mn or 1
    return ''.join(
        ticks[int((v-mn)/span*(len(ticks)-1))]
        for v in vals
    )

# 3) build the table with first/peak/last + sparkline
def build_sparkline_table(df3, terms):
    # pivot to have one row per term
    mini = df3.pivot(index='term', columns='stage', values='freq').loc[terms]
    out = []
    for term in terms:
        first, peak, last = mini.loc[term, ['first','peak','last']]
        out.append({
            'Term':  term,
            'First': f"{first:.1f}",
            'Peak':  f"{peak:.1f}",
            'Last':  f"{last:.1f}",
            'Trend': sparkline([first, peak, last]),
        })
    return pd.DataFrame(out)

# --- USAGE (in your notebook/script) ---
# 1) pick your top N terms
top_terms = trending.head(12)['term'].tolist()

# 2) build the 3-point df
df3, valid_terms = build_three_point_df(trending, peaks, top_terms)

# 3) build & display the table
table = build_sparkline_table(df3, valid_terms)
table
```
